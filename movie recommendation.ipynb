```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
import difflib #koi movie ka naam search krne pr.. spelling glt hone pr bhi kon sa movie uske
# closest hai vo pta lgane ke liye DIFFLIB ka use hota hai..
from sklearn.metrics.pairwise import cosine_similarity #it gives some similarity score between two values.
```

```python
df=pd.read_csv('movies.csv')
df2=pd.read_csv('movies.csv')
df.head()
```

```python
df.shape
```

```python
# selecting the relevant features
selected=['genres','id','keywords','original_title','cast','director','tagline','index']
```

```python
df=df[selected]
for i in selected:
    df[i]=df[i].fillna('')
```

```python
df.head()
```

```python
# combining all selected lists
combined_selection=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
combined_selection
```

```python
# feature extraction of text
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
combined_selection=feature_extraction.fit_transform(combined_selection)
```

```python
print(combined_selection)
```

```python
# getting the similarity score
similarity=cosine_similarity(combined_selection)
```

```python
print(similarity.shape)
```

```python
# list of all the movies
list_of_movies=df['original_title'].tolist()
```

```python
# input movie name from user
movie_name=input('enter the movie name: ')
```

```python
similar_movies=difflib.get_close_matches(movie_name,list_of_movies)
print(similar_movies)
```

```python
# finding index of the movie
k=df2[df2.original_title==similar_movies[0]]['index'].values[0]
print(k)
```

```python
# getting similar list of movies
similar_movies_index=list(enumerate(similarity[k])) #enumerate is used to avoide loop
# similarity score compared with input movie
```

```python
print(similar_movies_index)
```

```python
similar_movies_index=sorted(similar_movies_index,key=lambda x:x[1],reverse=True)
```

```python
print(similar_movies_index)
```

```python
movie_names=[]
for i in range(30):
    movie_names.append(df2[df2.index==similar_movies_index[i][0]]['title'].values[0])
```

```python
for i in movie_names:
    print('->'+ i)
```

```python
# incorporating all together
movie_name=input('enter the movie name: ')

similar_movies=difflib.get_close_matches(movie_name,list_of_movies)

k=df2[df2.original_title==similar_movies[0]]['index'].values[0]

similar_movies_index=list(enumerate(similarity[k]))

similar_movies_index=sorted(similar_movies_index,key=lambda x:x[1],reverse=True)

movie_names=[]
for i in range(40):
    movie_names.append(df2[df2.index==similar_movies_index[i][0]]['title'].values[0])
    
for i in movie_names:
    print('->'+ i)
```
